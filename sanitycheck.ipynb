```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
forcfile = '/glade/p/work/manab/fcast/MetSim/PNW/ensforcing20102016/ens_forc.PNW.GF.2010-2016.001.nc' # Orig forcing
forc = xr.open_dataset(forcfile).set_index(hru='hru')

metsimfile = '/glade/p/work/manab/fcast/MetSim/PNW/spinupforcing/20100601_20161231_3H/spinupforcing_20100601_20161231_3H.nc'
mets = xr.open_dataset(metsimfile)
plt.style.use('fivethirtyeight')
```

```python
ostart = '2010-06-01'
oend = '2016-12-31'

plt.figure(figsize=(14, 3))
forc.sel(hru = 17000001)['prcp'].loc[ostart:oend].plot()
plt.ylabel('Precip (mm)')
plt.title('HRU = 17000001, Daily Ensemble Forcing')
plt.show()

plt.figure(figsize=(14, 3))
forc.sel(hru = 17000001)['tmax'].loc[ostart:oend].plot()
plt.ylabel('tmax (C)')
plt.title('HRU = 17000001, Daily Ensemble Forcing')
plt.show()

plt.figure(figsize=(14, 3))
forc.sel(hru = 17000001)['tmin'].loc[ostart:oend].plot()
plt.ylabel('tmin (C)')
plt.title('HRU = 17000001, Daily Ensemble Forcing')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['pptrate'].plot()
plt.ylabel('Precip (kg m-2 s-1)')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['airtemp'].plot()
plt.ylabel('Air Temp (K)')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['airpres'].plot()
plt.ylabel('Air Pressure (Pa)')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['SWRadAtm'].plot()
plt.ylabel('SWRadAtm (W m-2)')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['LWRadAtm'].plot()
plt.ylabel('LWRadAtm (W m-2)')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()

plt.figure(figsize=(14, 3))
mets.sel(hru = 17000001)['spechum'].plot()
plt.ylabel('Spec. Humidity')
plt.title('HRU = 17000001, 3H Ensemble Forcing (MetSim)')
plt.show()
```
